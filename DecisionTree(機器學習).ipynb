```python
from sklearn import tree
from sklearn.model_selection import train_test_split
#連結資料庫(sklearn.datasets)設名稱為(load_iris)
from sklearn.datasets import load_iris

iris=load_iris()     #讀取資料
iris_data=iris.data    #資料
iris_label=iris.target  #分類代碼

#X_train(訓練資料)、X_test(測試資料)、y_train(訓練結果)、y_test(測試結果)
train_data,test_data,train_label,test_label=train_test_split(iris_data,iris_label,test_size=0.3)

# 建立分類器
#tree.DecisionTreeClassifier：分類樹
#tree.DecisionTreeRegressor：回歸樹
#tree.export_graphviz：將生成的決策樹導出為DOT格式，畫圖專用
#tree.ExtraTreeClassifier：高隨機版本的分類樹
#tree.ExtraTreeRegressor：高隨機版本的回歸樹

clf = tree.DecisionTreeClassifier()
iris_clf = clf.fit(train_data, train_label)

print("訓練資料正確率:",clf.score(train_data,train_label)) # 訓練資料正確率
print("測試資料正確率:",clf.score(test_data,test_label)) # 測試資料正確率
```

```python
#將上方結果存成pdf檔案
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("iris") 
```

```python
#將上方結果以圖形化呈現
import pydotplus  
from IPython.display import Image  
dot_data = tree.export_graphviz(clf, out_file=None, 
                feature_names=iris.feature_names,
                class_names=iris.target_names,  
                filled=True, rounded=True,  
                special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())  
```
